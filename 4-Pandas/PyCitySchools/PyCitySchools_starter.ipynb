```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset
scd = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
tot_schools = scd["School ID"].nunique()
tot_stu = scd["Student ID"].nunique()
tot_budget = scd["budget"].unique().sum()
avg_math = scd["math_score"].mean()
avg_read = scd["reading_score"].mean()
pass_math = len(scd[scd["math_score"] >=70])/tot_stu*100
pass_read = len(scd[scd["reading_score"] >=70])/tot_stu*100
pass_overall = (avg_math+avg_read)/2
```

```python
dis_sum = pd.DataFrame({"Total Schools": [tot_schools], "Total Students": [tot_stu], "Total Budget":[tot_budget], 
                        "Avg Math Score":[avg_math], "Avg Reading Score": [avg_read],"% Passing Math":[pass_math], 
                        "% Passing Reading": [pass_read], "% Passing Overall": [pass_overall]}, 
                       columns = ["Total Schools", "Total Students", "Total Budget", "Avg Math Score", "Avg Reading Score",
                                  "% Passing Math", "% Passing Reading", "% Passing Overall"])

dis_sum["Total Students"] = dis_sum["Total Students"].astype(int).map("{:,}".format)
dis_sum["Total Budget"] = dis_sum["Total Budget"].astype(float).map("${:,.0f}".format)
dis_sum["Avg Math Score"] = dis_sum["Avg Math Score"].astype(float).map("{:,.1f}%".format)
dis_sum["Avg Reading Score"] = dis_sum["Avg Reading Score"].astype(float).map("{:,.1f}%".format)
dis_sum["% Passing Math"] = dis_sum["% Passing Math"].astype(float).map("{:,.1f}%".format)
dis_sum["% Passing Reading"] = dis_sum["% Passing Reading"].astype(float).map("{:,.1f}%".format)
dis_sum["% Passing Overall"] = dis_sum["% Passing Overall"].astype(float).map("{:,.1f}%".format)

dis_sum
```

```python
scd_group = scd.groupby(["school_name"])
sc_type = scd_group["type"].last()
tot_stu = scd_group["Student ID"].count()
tot_budget = scd_group["budget"].last()
per_stu_budget = tot_budget/tot_stu
avg_math = scd_group["math_score"].mean()
avg_read = scd_group["reading_score"].mean()
pass_math = scd[scd["math_score"]>=70].groupby(["school_name"]).size()/tot_stu*100
pass_read = scd[scd["reading_score"]>=70].groupby(["school_name"]).size()/tot_stu*100
pass_overall = (pass_math+pass_read)/2
```

```python
school_sum_group_unformat = pd.DataFrame({"School Type": sc_type, "Total Students": tot_stu, "Total Budget": tot_budget, 
                        "Per Student Budget": per_stu_budget,
                        "Avg Math Score": avg_math, "Avg Reading Score": avg_read,"% Passing Math": pass_math, 
                        "% Passing Reading": pass_read, "% Passing Overall": pass_overall}, 
                       columns = ["School Type","Total Students", "Per Student Budget", "Avg Math Score", "Avg Reading Score",
                                  "% Passing Math", "% Passing Reading", "% Passing Overall"])
```

```python
school_sum_group = pd.DataFrame({"School Type": sc_type, "Total Students": tot_stu, "Total Budget": tot_budget, 
                        "Per Student Budget": per_stu_budget,
                        "Avg Math Score": avg_math, "Avg Reading Score": avg_read,"% Passing Math": pass_math, 
                        "% Passing Reading": pass_read, "% Passing Overall": pass_overall}, 
                       columns = ["School Type","Total Students", "Per Student Budget", "Avg Math Score", "Avg Reading Score",
                                  "% Passing Math", "% Passing Reading", "% Passing Overall"])

school_sum_group["Total Students"] = school_sum_group["Total Students"].astype(int).map("{:,}".format)
school_sum_group["Per Student Budget"] = school_sum_group["Per Student Budget"].astype(float).map("${:,.0f}".format)
school_sum_group["Avg Math Score"] = school_sum_group["Avg Math Score"].astype(float).map("{:,.1f}".format)
school_sum_group["Avg Reading Score"] = school_sum_group["Avg Reading Score"].astype(float).map("{:,.1f}".format)
school_sum_group["% Passing Math"] = school_sum_group["% Passing Math"].astype(float).map("{:,.1f}%".format)
school_sum_group["% Passing Reading"] = school_sum_group["% Passing Reading"].astype(float).map("{:,.1f}%".format)
school_sum_group["% Passing Overall"] = school_sum_group["% Passing Overall"].astype(float).map("{:,.1f}%".format)


school_sum_group
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_9 = scd.loc[scd["grade"] == "9th"].groupby("school_name")["math_score"].mean()
math_10 = scd.loc[scd["grade"] == "10th"].groupby("school_name")["math_score"].mean()
math_11 = scd.loc[scd["grade"] == "11th"].groupby("school_name")["math_score"].mean()
math_12 = scd.loc[scd["grade"] == "12th"].groupby("school_name")["math_score"].mean()

math_avg_grade = pd.DataFrame({"9th": math_9, "10th": math_10, "11th": math_11, "12th": math_12})

math_avg_grade["9th"] = math_avg_grade["9th"].astype(float).map("{:,.1f}".format)
math_avg_grade["10th"] = math_avg_grade["10th"].astype(float).map("{:,.1f}".format)
math_avg_grade["11th"] = math_avg_grade["11th"].astype(float).map("{:,.1f}".format)
math_avg_grade["12th"] = math_avg_grade["12th"].astype(float).map("{:,.1f}".format)


math_avg_grade
```

## Scores by School Spending

  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 615, 645, 675]
labels = ["0 to 585", "585 to 615", "615 to 645", "645 to 675"]
```
